```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("hepatitis_csv.csv")
```

## Description

Hepatitis is a liver disease which mainly attacks the liver and can be fatal. From 2000 to 2016 there were 15916 reported cases.<sup>1</sup>

The used dataset for this  consists  of 155 patient entries and 20 feature taken from a public kaggle repository <sup>2</sup>

<sub>1 https://www.ages.at/service/service-presse/pressemeldungen/hepatitis-in-oesterreich/</sub>

<sub> 2 https://www.kaggle.com/codebreaker619/hepatitis-data</sub>

```python
continues= np.array(["age","bilirubin","alk_phosphate","sgot","albumin","protime"])
categorical = np.array(["sex","steroid","antivirals","fatigue","malaise","anorexia","liver_big","spleen_palpable","spiders","ascites","varices","histology","class"])
data[continues] = data[continues].replace("NaN",np.NAN)
data[categorical] = data[categorical].replace("NaN",np.NAN)
data.head()
```

### Continues Features

* Age 
* Bilirubin - A side product of the destruction of red blood cells which usually is filtered out by the liver
* alk_phosphate
* sgot - Serum Glutamic-Oxaloacetic transaminase(SGOT) also known as AST (aspartate aminotransferase) is a enzym whinin the liver which leaks into the bloodstream when cells are damaged.
* albumin - Albumin is a protein made by the liver which prevents fluid from leaking out of blood vessels into tissues 
* protime - Prothrombin, a protein which helps the blood to clot. The prothrombin time (PT or protime) measures how long it takes for the blood to clot. 

```python
print(data.describe())
fig = data.hist(column=continues, figsize=(12,8))
```

### Categorical Features


```python
count=1
plt.subplots(figsize=(20, 40))
for c in categorical:
    plt.subplot(7,2,count)
    sns.countplot(x=c,data=data).set_title(c)
    count+=1

plt.show()
```
